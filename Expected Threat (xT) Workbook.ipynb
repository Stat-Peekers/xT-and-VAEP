# Computing the xT Grid
$xT_{x,y}=(s_{x,y} \times g_{x,y})+(m_{x,y} \times \displaystyle\sum_{z=1} ^{16}\sum_{w=1} ^{12} T_{(x,y)->(z,w)} \times xT_{z,w}))$

```python
move_actions = ["Pass", "TakeOn"]
shoot_actions = ['SavedShot', 'MissedShots', 'ShotOnPost', 'Goal']
ms_actions = move_actions + shoot_actions
```

```python
xt_df[(xt_df["type"] == "Goal") & (xt_df["isOwnGoal"] == True)].shape
```

```python
df1 = xt_df.groupby(["x1_bin", "y1_bin"]).agg({"type": lambda x: x.isin(move_actions).sum()})
df2 = xt_df.groupby(["x1_bin", "y1_bin"]).apply(lambda x: (x["isShot"] & (x["isOwnGoal"] != True)).sum())
df3 = xt_df.groupby(["x1_bin", "y1_bin"]).apply(lambda x: (x["isGoal"] & (x["isOwnGoal"] != True)).sum())
ms_df = pd.concat([df1, df2, df3], axis=1)
del df1, df2, df3
```

```python
ms_df.head()
```

```python
ms_df.columns = ["move_actions", "shoot_actions", "goals"]
```

```python
ms_df.reset_index().head()
```

```python
ms_df["ms_actions"] = ms_df["move_actions"].add(ms_df["shoot_actions"])
```

```python
ms_df["move_prob"] = ms_df["move_actions"].divide(ms_df["ms_actions"])
ms_df["shoot_prob"] = ms_df["shoot_actions"].divide(ms_df["ms_actions"])
```

```python
start_time = time.clock()
xt_cols=16; xt_rows=12
tot_iterations = 5
xT_zw = pd.DataFrame(np.zeros(goal_matrix.shape))
for n_iter in range(tot_iterations):
    print("Iteration no.", n_iter+1)
    t_xy_zw_final = pd.DataFrame(np.zeros([xt_rows, xt_cols]), index=range(xt_rows), columns=range(xt_cols))
    for z in range(xt_cols):
        for w in range(xt_rows):
            df = xt_df[(xt_df["end_xy"] == (z, w))].groupby(["x1_bin", "y1_bin"]).agg({"type": lambda x: x.isin(move_actions).sum()}).reset_index()
            df["move_prob"] = df["type"].divide(df["type"].sum())
            t_xy_zw = pd.DataFrame(df.pivot_table(values="move_prob", index="y1_bin", columns="x1_bin"),
                                   index=range(xt_rows), columns=range(xt_cols))
            t_xy_zw_final = t_xy_zw_final + (t_xy_zw * xT_zw.iloc[w,z]).fillna(0)
    print(t_xy_zw_final)
    xT_final = (shoot_matrix*goal_matrix)+(move_matrix*t_xy_zw_final)
    print(xT_final)
    xT_zw = xT_final
stop_time = time.clock()
print(stop_time-start_time)
```
